```python
from flask import Flask, jsonify
```

```python
# Jupyter Notebook içerisinde resim göstermek için import edildi
from IPython.display import Image
```

```python
app = Flask(__name__)
```

```python
movies = [{'id': '0',
           'name': 'The Godfather',
           'director': 'Francis Ford Coppola',
           'date': 1972},
         {'id': '1',
           'name': 'The Shawshank Redemption',
           'director': 'Frank Darabont',
           'date': 1994},
         {'id': '2',
           'name': 'The Matrix',
           'director': 'Lana Wachowski - Lilly Wachowski',
           'date': 1999},
         {'id': '3',
           'name': 'The Dark Night',
           'director': 'Christopher Nolan',
           'date': 2008},
         {'id': '4',
           'name': 'Inception',
           'director': 'Christopher Nolan',
           'date': 2010},
         {'id': '5',
           'name': 'Interstellar',
           'director': 'Christopher Nolan',
           'date': 2014}
         ]
```

```python
@app.route('/')
def index():
    return "Welcome To The Movies API"
```

```python
Image(filename='api_test_1.jpg')
```

```python
@app.route('/movies', methods=['GET'])
def get():
    return jsonify({'Movies': movies})
```

```python
Image(filename='api_test_2.jpg')
```

```python
@app.route('/movies/<int:id>', methods=['GET'])
def get_movie(id):
    return jsonify({'Movie': movies[id]})
```

```python
if __name__ == "__main__":
    app.run()
```
