```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Yeni klasor\\Tesseract-OCR\\tesseract.exe'
img_1 = cv2.imread('cr1.jpg')
img_2 = cv2.imread('cr2.jpg')
img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
img_2 = cv2.cvtColor(img_2, cv2.COLOR_BGR2RGB)
```

```python
print(pytesseract.image_to_string(img_1))
```

```python
print(pytesseract.image_to_string(img_2))
```

```python
# Kelimeleri Tanımlama

hImg, wImg, _ = img_1.shape
boxes = pytesseract.image_to_data(img_1)
for x, b in enumerate(boxes.splitlines()):
    if x != 0:
        b = b.split()
        #print(b)
        if len(b) == 12:
            x, y, w, h = int(b[6]), int(b[7]), int(b[8]), int(b[9])
            cv2.rectangle(img_1, (x, y), (w + x, h + y), (0, 0, 255), 1)
            cv2.putText(img_1, b[11], (x, y - 1), cv2.FONT_HERSHEY_COMPLEX, 0.4, (50, 50, 255), 1)
```

```python
plt.figure(figsize=(32, 24), dpi=80)
plt.imshow(img_1)
```
