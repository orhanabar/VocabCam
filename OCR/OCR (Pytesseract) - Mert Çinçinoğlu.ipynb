```python
import cv2
import pytesseract
from pdf2image import convert_from_path
#import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:\\Yeni klasor\\Tesseract-OCR\\tesseract.exe'
```

```python
def Print_File_Content(file_name):
    file_str = ''

    if file_name.endswith('.pdf'):
        # Store Pdf with convert_from_path function
        pdf_file = convert_from_path(file_name)

        for i in range(len(pdf_file)):
            # Save pages as images in the pdf
            pdf_file[i].save('page' + str(i) + '.jpg', 'JPEG')
            # TODO: Create a different folder for the outputs (try / except) and delete after.
            # TODO: Or do it without save.
            file = 'page' + str(i) + '.jpg' # TODO: Add images taken from pdf to list.
            img = cv2.imread(file)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            file_str += pytesseract.image_to_string(img)
    elif file_name.endswith('.jpg'):
        file = file_name
        # TODO: It should be able to take more than one picture in a list.
        img = cv2.imread(file)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        file_str = pytesseract.image_to_string(img)
        
    print(file_str)
```

```python
Print_File_Content('Computer Reading.pdf')
```
